```python
import numpy as np
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
df = pd.read_csv('Demographic_details_Country.CSV')
df.head()
```

```python
pd.set_option('display.max_rows', None)
df
```

```python
#Clean & Inspect Data mate
df.info()
df.describe()
df.isnull().sum()
```

```python
df.dropna(inplace=True)
```

```python
df.sort_values(by='Total revenue', ascending=False)
```

```python
#kay country ma vadhare revenue chhe

top_revenue = df.sort_values(by='Total revenue', ascending=False).head(25)
plt.figure(figsize=(15, 10))
sns.barplot(
    x='Total revenue',
    y='Country',
    hue='Country',
    data=top_revenue,
    palette='BuPu',
    legend=False
)
plt.title('Top 25 Countries by Total revenue')
plt.xlabel('Total revenue')
plt.ylabel('Country')
plt.tight_layout()
plt.show()
```

```python
#kay country ma vadhare active user chhe

top_active = df.sort_values(by='Active users', ascending=False).head(25)
plt.figure(figsize=(15, 10))
sns.barplot(
    x='Active users',
    y='Country',
    hue='Country',
    data=top_active,
    palette='Blues_d',
    legend=False
)
plt.title('Top 25 Countries by Active Users')
plt.xlabel('Active Users')
plt.ylabel('Country')
plt.tight_layout()
plt.show()
```

```python
#kay country ma vadhare New user aavya chhe

top_new = df.sort_values(by='New users', ascending=False).head(25)
plt.figure(figsize=(15, 10))
sns.barplot(
    x='New users',
    y='Country',
    hue='Country',
    data=top_new,
    palette='crest',
    legend=False
)
plt.title('Top 25 Countries by New Users')
plt.xlabel('New Users')
plt.ylabel('Country')
plt.tight_layout()
plt.show()
```

```python
#Top 25 countries revenue
top_rev = df.sort_values('Total revenue', ascending=False).head(25)

plt.figure(figsize=(15, 10))
sns.barplot(x='Total revenue', y='Country', data=top_rev)
plt.title('Top 25 Countries by Revenue')
plt.xlabel('Revenue')
plt.ylabel('Country')
plt.xticks(np.arange(0.0, 0.32, 0.01))
plt.tight_layout()
plt.show()
```

```python
#Monthly active use (MAU)

start_date = pd.to_datetime("2025-05-28")
end_date = pd.to_datetime("2025-06-24")
days_range = (end_date - start_date).days

df['MAU_estimate'] = df['Active users']
```

```python
top_dau = df.sort_values('MAU_estimate', ascending=False).head(25)
top_dau['Country Group'] = 'Top 25 MAU'
plt.figure(figsize=(15, 10))
sns.barplot( x='MAU_estimate', y='Country', data=top_dau, hue='Country Group', palette='YlGnBu')
plt.title('Top 10 Countries by Estimated Monthly Active Users (MAU)')
plt.xlabel('Estimated MAU')
plt.ylabel('Country')
plt.legend(title='Monthly active user', loc='center right')
plt.show()
```

```python
#Average engagement time per active user (kya contry na user ae ketlo time aahi engage rhe chhe)
# sheet ma sec ma batavyu chhe jene pehla min ma convert karyu

df['Avg_engage_min'] = df['Average engagement time per active user'] / 60

top_session = df.sort_values('Avg_engage_min', ascending=False).head(25)

# Step 3: Plot
plt.figure(figsize=(18, 10))
sns.barplot(
    x='Avg_engage_min',
    y='Country',
    data=top_session,
    hue='Country',
    palette='crest',
    legend=False 
)
plt.title('Top 25 Countries by Average Session Duration (in minutes)')
plt.xlabel('Avg Session Duration (minutes)')
plt.ylabel('Country')

max_minutes = round(top_session['Avg_engage_min'].max()) + 5
plt.xticks(np.arange(0, max_minutes, 5))

plt.tight_layout()
plt.show()
```

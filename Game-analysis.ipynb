```python
import numpy as np
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
```

```python

df = pd.read_csv('Demographic_details_Country.CSV')
df
```

```python
df.info()
df.describe()
df.isnull().sum()
```

```python
df.dropna(inplace=True)
```

```python
df.sort_values(by='Total revenue', ascending=False)
```

```python
top_revenue = df.sort_values(by='Total revenue', ascending=False).head(25)
plt.figure(figsize=(15, 10))
sns.barplot(
    x='Total revenue',
    y='Country',
    hue='Country',
    data=top_revenue,
    palette='BuPu',
    legend=False
)
plt.title('Top 25 Countries by Total revenue')
plt.xlabel('Total revenue')
plt.ylabel('Country')
plt.tight_layout()
plt.show()
```

```python
top_active = df.sort_values(by='Active users', ascending=False).head(25)
plt.figure(figsize=(15, 10))
sns.barplot(
    x='Active users',
    y='Country',
    hue='Country',
    data=top_active,
    palette='Blues_d',
    legend=False
)
plt.title('Top 25 Countries by Active Users')
plt.xlabel('Active Users')
plt.ylabel('Country')
plt.tight_layout()
plt.show()
```

```python
top_new = df.sort_values(by='New users', ascending=False).head(25)
plt.figure(figsize=(15, 10))
sns.barplot(
    x='New users',
    y='Country',
    hue='Country',
    data=top_new,
    palette='crest',
    legend=False
)
plt.title('Top 25 Countries by New Users')
plt.xlabel('New Users')
plt.ylabel('Country')
plt.tight_layout()
plt.show()
```

```python
fig, ax1 = plt.subplots(figsize=(20, 6))
x = np.arange(len(df['Country']))

ax1.bar(x, df['Engaged sessions per active user'], color='skyblue')
ax1.set_ylabel('Engaged Sessions per User', color='skyblue')
ax1.tick_params(axis='y', labelcolor='skyblue')

ax2 = ax1.twinx()
ax2.plot(x, df['Average engagement time per active user'], color='orange', marker='o')
ax2.set_ylabel('Avg Engagement Time (sec)', color='orange')
ax2.tick_params(axis='y', labelcolor='orange')

ax1.set_xticks(x)
ax1.set_xticklabels(df['Country'], rotation=60, ha='right', fontsize=8)
ax1.set_xlabel('Country')

ax2.legend(['Avg Engagement Time (sec)', 'Engaged Sessions per User'], loc='upper left', bbox_to_anchor=(0, 1.15))

plt.title('Country-wise Engagement: Sessions vs Avg Time')
fig.tight_layout()
plt.show()

```

```python
avg_sessions = df['Engaged sessions per active user'].mean()
avg_time = df['Average engagement time per active user'].mean()

filtered_df = df[
    (df['Engaged sessions per active user'] > avg_sessions) &
    (df['Average engagement time per active user'] > avg_time)
]

filtered_df = filtered_df.sort_values(by='Engaged sessions per active user', ascending=False).reset_index(drop=True)

fig, ax1 = plt.subplots(figsize=(20, 6))
x = np.arange(len(filtered_df['Country']))

ax1.bar(x, filtered_df['Engaged sessions per active user'], color='skyblue')
ax1.set_ylabel('Engaged Sessions per User', color='skyblue')
ax1.tick_params(axis='y', labelcolor='skyblue')

ax2 = ax1.twinx()
ax2.plot(x, filtered_df['Average engagement time per active user'], color='orange', marker='o')
ax2.set_ylabel('Avg Engagement Time (sec)', color='orange')
ax2.tick_params(axis='y', labelcolor='orange')

ax1.set_xticks(x)
ax1.set_xticklabels(filtered_df['Country'], rotation=60, ha='right', fontsize=8)
ax1.set_xlabel('Country')

ax2.legend(['Avg Engagement Time (sec)', 'Engaged Sessions per User'], loc='upper left', bbox_to_anchor=(0, 1.15))

plt.title('Top Countries: High Sessions & Engagement Time')
fig.tight_layout()
plt.show()
```
